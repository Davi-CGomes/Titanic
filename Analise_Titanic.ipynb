## Realizando os tratamentos iniciais dos dados

```python
treino.isnull().sum().sort_values(ascending=False).head(5)
```

```python
# Retornando o shape (dimensão da tabela = linhas, colunas)
treino.shape
```

```python
treino.head(4)
```

```python
# Verificando a cardinalidade (quantiade de valores únicos em uma coluna) dos dados.
treino.nunique().sort_values(ascending=False)
```

**Observação:** Colunas com alta cardinalidade como nome, número do ticket e até mesmo o código da cabine não vão ajudar o modelo pois podem torná-los menos generalizável. Nesse caso, elimina essas colunas.

```python
# Eliminando para base de treino
treino = treino.drop(['Name', 'Ticket', 'Cabin'], axis=1)
```

```python
# Eliminando, também, na base de teste
teste = teste.drop(['Name', 'Ticket', 'Cabin'], axis=1)
```

**Observação:** Precisa tratar a coluna Age, que possui valores vazios tanto no treino quanto no teste. Como essa coluna possui a idade dos passageiros, utiliza-se a média das idades para substituir os valores vazios.

```python
# Verificando a média das idades para a base de treino
treino.Age.mean()
```

```python
# Selecionando as colunas com idade vazia e substituindo pela média
treino.loc[treino.Age.isnull(),'Age'] = treino.Age.mean()
```

```python
# O mesmo para a base de teste
teste.loc[teste.Age.isnull(),'Age'] = teste.Age.mean()
```

**Observação:** Na base de treino, ainda tem 2 valores vazios na coluna Embarked. Verificar todos os valores nessa coluna e substituir os valores vazios pela moda.

```python
# Verificando os valores na coluna Embarked
treino.Embarked.value_counts()
```

```python
# Selecionando a moda dessa coluna
treino.Embarked.mode()[0]
```

```python
# Substituindo os valores vazios
treino.loc[treino.Embarked.isnull(),'Embarked'] = treino.Embarked.mode()[0]
```

**Observação:** A coluna Fare ainda tem 1 valor vazio. Pode ser usado a mesma ideia que foi feito para a coluna Age e substituindo pela moda.

```python
# Verificando a média da coluna fare para a base de teste
teste.loc[teste.Fare.isnull(),'Fare'] = teste.Fare.mean()
```

## Considerando apenas as colunas que não são de texto

```python
# Verificando as colunas de texto
col_treino_nr = treino.columns[treino.dtypes != 'object']
col_treino_nr
```

```python
# Selecionando apenas os valores númericos da base de treino
treino_nr = treino.loc[:,col_treino_nr]
```

```python
# Na base de teste
col_teste_nr = teste.columns[teste.dtypes != 'object']
col_teste_nr
```

```python
# Valores númericos da base de teste
teste_nr = teste.loc[:,col_teste_nr]
```

## Testando modelos para classificar esses dados

### Separando a base de treino em treino e validação usando o train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
# Separando a base de treino em x e y
X = treino_nr.drop(['PassengerId', 'Survived'], axis=1)
y = treino.Survived 
```

```python
# Separando em treino e validação
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.33, random_state=42)
```

### Usando a árvore de classificação

**Observação:**  prevê a categoria (classe) de um dado através de uma série de perguntas "se-então". </br>
https://scikit-learn.org/stable/modules/tree.html

```python
# Fazendo a importação
from sklearn import tree
```

```python
# Criando o classificador
clf_ac = tree.DecisionTreeClassifier(random_state=42)
```

```python
# Fazendo fit com os dados
clf_ac = clf_ac.fit(X_train, y_train)
```

```python
# Fazendo a previsão
y_pred_ac = clf_ac.predict(X_val)
```

### Usando KNeighborsClassifier

**Observação:** KNN é um algoritmo que classifica um novo dado com base nos rótulos dos K exemplos mais próximos a ele. </br> https://scikit-learn.org/stable/modules/neighbors.html


```python
# Importando
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Criando o classificador
clf_knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# Fazendo o fit com os dados
clf_knn = clf_knn.fit(X_train, y_train)
```

```python
# Fazendo a previsão
y_pred_knn = clf_knn.predict(X_val)
```

### Usando a Regressão Logistica 

**Observação:** Regressão logística funciona calculando uma combinação dos dados de entrada e transformando isso em uma probabilidade entre 0 e 1 para decidir a classe. </br>https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html

```python
# Importando
from sklearn.linear_model import LogisticRegression
```

```python
# Criando o classificador
clf_r1 = LogisticRegression(random_state=42)
```

```python
# Fazendo o fit com os dados
clf_r1 = clf_r1.fit(X_train, y_train)
```

```python
# Fazendo a previsão
y_pred_r1 = clf_r1.predict(X_val)
```

## Avaliando modelos

### Acurácia

```python
# Importando
from sklearn.metrics import accuracy_score
```

```python
# Para a árvore
accuracy_score(y_val, y_pred_ac)
```

```python
# Para o Knn
accuracy_score(y_val, y_pred_knn)
```

```python
# Para a Regressão Logística
accuracy_score(y_val, y_pred_r1)
```

### Avaliando a matriz de confunsão 
**Tabela que compara as previsões do modelo com os valores reais, mostrando acertos e erros por classe.**

```python
# Importando
from sklearn.metrics import confusion_matrix
```

```python
# Para a árvore
confusion_matrix(y_val, y_pred_ac)
```

```python
# Para o Knn
confusion_matrix(y_val, y_pred_knn)
```

```python
# Para a Regressão Logística
confusion_matrix(y_val, y_pred_r1)
```
